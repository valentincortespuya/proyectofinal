```python
import PyPDF2

def extraer_texto_de_pdf(archivo_pdf):
    try:
        # Abre el archivo PDF en modo lectura binaria
        with open(archivo_pdf, 'rb') as pdf_file:
            # Crea un objeto PdfReader para el archivo PDF
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            
            # Inicializa una cadena para almacenar el texto extraído
            texto = ''
            
            # Itera a través de cada página del PDF y extrae el texto
            for pagina in pdf_reader.pages:
                texto += pagina.extract_text()
            
            return texto
    except Exception as e:
        print(f"Error al extraer texto del PDF: {str(e)}")
        return None

# Ejemplo de uso
archivo_pdf = 'document.pdf'
texto_extraido = extraer_texto_de_pdf(archivo_pdf)

if texto_extraido:
    print(texto_extraido)
else:
    print("No se pudo extraer texto del PDF.")
```

```python
import re



# Dividir el texto en líneas
lineas = texto_extraido.split('\n')

# Filtrar las líneas que comienzan con una fecha en formato "dd-mm-aaaa"
filas_con_fecha = [linea.strip() for linea in lineas if re.match(r'\d{2}-\d{2}-\d{4}', linea)]

# Eliminar todo después del segundo € en cada fila
for i, fila in enumerate(filas_con_fecha):
    match = re.search(r'€[^€]*€', fila)
    if match:
        filas_con_fecha[i] = fila.replace(match.group(0), match.group(0)[:-1])

# Imprimir las filas resultantes
for fila in filas_con_fecha:
    print(fila)
```
